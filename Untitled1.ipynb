# Using ACO to solve the TSP problem

The evolutionary algorithm is one way to solve a complex problem with an exponential amount of solutions, but there also exist another suitable algorithm for the TSP.

The Ant Colony Optimization, or ACO, is an algorithm based on how a colony of ants traverses through multiple points.

Ants leave pheromones whenever they travel for other ants to follow.  The other ants figure out faster ways how to travel through a group of points based on the pheremones they followed through, then more and more groups of ants figure out faster and faster ways to travel.

The great thing about AI ants as opposed to real ants is that since AIs has memories, they know not to visit the points they already visited.  And they already know the distances between the two and tend to choose the next nearest point.  If the distances between two points are the same, the AI and will choose the path with the most pheremones (just like a real ant).

When calling on the ACO file, the user has a choice between two strategies:

(1) The ant quality system

or

(2) The ant Quantity system

```python
def _update_pheromone_delta(self):
        self.pheromone_delta = [[0 for j in range(self.graph.rank)] for i in range(self.graph.rank)]
        for _ in range(1, len(self.tabu)):
            i = self.tabu[_ - 1]
            j = self.tabu[_]
            if self.colony.update_strategy == 1: 
                # ant-quality system: Increase Trail independent of pheremone delta 
                self.pheromone_delta[i][j] = self.colony.Q
            # ant-density system Shorter edges made more desirable by making trails inversely proportional to pheremone delta
            elif self.colony.update_strategy == 2:  
                # noinspection PyTypeChecker
                self.pheromone_delta[i][j] = self.colony.Q / self.graph.matrix[i][j]
            else:  # ant-cycle system
                self.pheromone_delta[i][j] = self.colony.Q / self.total_cost
```
